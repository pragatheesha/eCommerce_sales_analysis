```python
#Bar Chart

import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Category": ["Electronics", "Accessories", "Clothing"],
    "Sales": [1500, 450, 300]
}

df = pd.DataFrame(data)

plt.figure()
plt.bar(df["Category"], df["Sales"])
plt.title("Sales by Product Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

```python
#Line Chart

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
temperature = [18, 20, 25, 30, 35, 38]

plt.figure()
plt.plot(months, temperature, marker='o')
plt.title("Monthly Temperature Changes")
plt.xlabel("Month")
plt.ylabel("Temperature (Â°C)")
plt.grid(True)
plt.show()
```

```python
#Pie Chart

expenses = {
    "Rent": 40,
    "Food": 25,
    "Transport": 15,
    "Entertainment": 10,
    "Savings": 10
}

labels = expenses.keys()
values = expenses.values()

plt.figure()
plt.pie(values, labels=labels, autopct="%1.1f%%")
plt.title("Monthly Expense Distribution")
plt.show()
```

```python
#Combine Data Loading, Cleaning, Analysis & Visualization

import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("/content/sales_data.csv")
df

# Clean data
df.dropna(inplace=True)

# Analyze
category_sales = df.groupby("Region")["Total_Sales"].sum()

# Visualize
plt.figure()
category_sales.plot(kind="bar")
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Revenue")
plt.show()
```
